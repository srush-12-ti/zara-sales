```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


zarad = pd.read_csv(r"C:\Users\srush\OneDrive\Desktop\datascience\zara.csv", sep=';')
data = pd.DataFrame(zarad)
#plotting countplot to get the sales volume range
sns.countplot(x='Sales Volume', data=data)
plt.xticks(rotation=10)
plt.title("Sales Volume Distribution")

plt.show()

#to_numeric converts the volume to the numeric value
data['Sales Volume']= pd.to_numeric(data['Sales Volume'])


high_sales = data[data['Sales Volume'] > 1000]


if not high_sales.empty:
    print("\nProducts with Sales Volume > 1000:\n")
    print(high_sales[['Product ID', 'Sales Volume','price','Promotion']])
else:
    print("No product has Sales Volume over 1000.")
```
